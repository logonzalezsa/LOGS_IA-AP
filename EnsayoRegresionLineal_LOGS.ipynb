## 1. INTRODUCCIÓN

En estadística, se conoce como análisis de regresión al proceso estadístico para estimar las relaciones entre variables, donde una de ellas es considerada como dependiente, mientras que las otras son consideradas como independientes, es decir que dicho proceso estadístico busca encontrar la relación entre la variable dependiente y las variables independientes. En otras palabras, se busca un modelo matemático que muestre como la variable dependiente es influenciada por las variables independientes. Frecuentemente, la variable dependiente es representada por Y, mientras que la notación Xi es usada para las variables independientes, las cuales pueden estar en una cantidad m, siendo m un número entero positivo. [Wikipedia](https://es.wikipedia.org/wiki/An%C3%A1lisis_de_la_regresi%C3%B3n)

En el análisis de regresión, es de interés además de definir una función para las variables independientes, que es conocida como función de regresión, también caracterizar la variación de la variable dependiente en torno a la función de regresión, que puede ser descrita por una función de probabilidad. ([Tusell (2011)](http://www.et.bs.ehu.es/~etptupaf/nuevo/ficheros/estad3/nreg1.pdf))

El uso del análisis de regresión comprende diversas finalidades como [Wikipedia](https://es.wikipedia.org/wiki/An%C3%A1lisis_de_la_regresi%C3%B3n):

* Predicción y previsión, donde su uso tiene superposición sustancial en el campo de aprendizaje automático.

* Entendimiento y razonamiento de cuales variables independientes están relacionadas con la variable dependiente.

* Inferencia de relaciones casuales entre las variables dependiente y las independientes. Sin embargo, es importante señalar que la correlación no implica casualidad ([Armstrong (2012)](https://www.academia.edu/1105970/Illusions_in_Regression_Analysis))

Diversas técnicas se han desarrollado para realizar el análisis de regresión, pero estas se agrupan en dos grandes técnicas, la regresión paramétrica y la regresión no paramétrica. La regresión paramétrica es aquella en la cual la función de regresión se define en términos de un número finito de parámetros desconocidos que se estiman a partir de los datos, y dentro de estos métodos son de uso familiar la regresión lineal y la regresión por cuadrados mínimos ordinarios ([Tusell (2011)](http://www.et.bs.ehu.es/~etptupaf/nuevo/ficheros/estad3/nreg1.pdf)).Por su parte, la regresión no paramétrica corresponde a aquella que permiten que la función de regresión consista en un conjunto de funciones de dimensión infinita ([Servy, García y Paccapelo (2006)](http://m.rephip.unr.edu.ar/bitstream/handle/2133/8791/Servy%20y%20Garcia%20regresi%C3%B3n%20no%20param%C3%A9trica.pdf?sequence=3&isAllowed=y)).

En el análisis de regresión, es importante también el desempeño del método de análisis, el cual depende de la forma del proceso de generación de datos, y cómo se relaciona con el método de regresión que se utiliza. Con frecuencia se inicia con un desconocimiento de la forma verdadera de generación de los datos y la función de regresión depende de hacer suposiciones acerca del proceso ([Freedman (2009)](https://pdfroom.com/books/statistical-models-theory-and-practice/OX623lXL54Z)).

Como se mencionó dentro de las técnicas de análisis de regresión, una de las más frecuentes es la regresión lineal. La regresión lineal es un modelo matemático también conocido como ajuste lineal usado para aproximar la relación de dependencia entre una variable dependiente *Y*, *m* variables independientes *Xi* con *m* entero positivo y un término aleatorio *épsilon*, el cual puede ser expresado como:

$Y = B0 + B1 X1 + ... + Bm Xm + épsilon$

Donde,

*Y* es la variable dependiente.

*X1*, *X2*, ... , *Xm* son las variables explicativas, independientes o regresoras.

*B0*, *B1*, *B2*, ..., *Bm* son los parámetros del modelo, y miden la influencia que las variables explicativas tienen sobre la regresión. *B0* es conocido como la intersección o término constante, mientras que las *Bi* (*i* = 1, 2, ... *m*) son los parámetros respectivos a cada variable independiente, y *m* es el número de parámetros independientes a tener en cuenta en la regresión; la regresión puede ser contrastada con la regresión no lineal.

La figura 1, muestra la representación gráfica de una función de regresión lineal con una variable dependiente y una variable independiente. Este tipo de función se conoce como Regresión Lineal Simple, para diferenciar de la Regresión Lineal Multivariable, en la cual más de dos variables independientes son involucradas en el modelo matemámtico.

**Figura 1: Representación gráfica de una función de regresión lineal simple.**

Por su parte, la regresión no lineal es un modelo de inferencia del tipo $Y = f(X, Teta) + épsilon$, donde f es una función no lineal respecto a algunos parámetros desconocidos Teta. Como mínimo, se pretende obtener valores de los parámetros asociados con la mejor curva de ajuste, que habitualmente se hace con el método de los mínimos cuadrados. Generalmente, se determina si el modelo es adecuado a partir del uso de conceptos de inferencia estadística como por ejemplo, intervalos de confianza para los parámetros y pruebas de bondad del ajuste.

La figura 2, muestra la representación gráfica de una función de regresión lineal, en la cual también un variable dependiente es un función de una variable independiente, sin embargo, ésta variable independiente puede estar involucrada en el modelo matemática en más de una forma. Por ejemplo, $X^n$,con *n* un número real diferente de cero, entre otras tipologías de no linealidad (uso de logaritmos, exponenciales, funciones trinogométricas).

**Figura 2: Representación grafica de una función de regresión no lineal.**

## 2. RESEÑA HISTÓRICA

El método de los mínimos cuadrados fue la primera forma de regresión, publicado en 1805 por Legendre ([Legendre, 1805](https://books.google.ca/books?id=FRcOAAAAQAAJ&pg=PA1&hl=es&source=gbs_toc_r&cad=3#v=onepage&q&f=false)), y en 1809 por Gauss ([Gauss, 1809](https://archive.org/details/bub_gb_ORUOAAAAQAAJ/page/n3/mode/2up)). Ellos, a partir de observaciones astronómicas, usaron el método para determinar las órbitas de los cuerpos alrededor del sol (inicialmente cometas, pero posteriormente usado para los recién descubiertos planetas menores). En 1821, con la inclusión de una versión del Teorema de Gauss-Markov, Gauss publicó un desarrollo posterior al método de los mínimos cuadrados ([Gauss, 1821](https://books.google.co.za/books?id=ZQ8OAAAAQAAJ&printsec=frontcover&dq=Theoria+combinationis+observationum+erroribus+minimis+obnoxiae&as_brr=3&hl=es#v=onepage&q&f=false)).

Para la descripción de un fenómeno biológico, Galton acuñó el término *REGRESIÓN* [Galton, 1877](https://galton.org/essays/1870-1879/galton-1877-typical-laws-heredity.pdf). Aunque la utilización de Galton se enmarcó en el ámbito biológico, Yule [Yule, 1897](https://www.jstor.org/stable/2979746?origin=crossref&seq=1) y Pearson et al. [Pearson; Yule; Blanchard; Lee, 1903](https://academic.oup.com/biomet/article/2/2/211/207458), extendieron el término de regresión, en el contexto estadístico, para quienes la distribución conjunta de la variable de respuesta y las variables explicativas es Gaussiana.

Sin embargo, en la década de los 20's del siglo XX, la anterior concepción fue refutada de tal manera, que se supuso que la distribución condicional de la respuesta era Gaussiana, pero la distribución conjunta no necesariamente lo fuera, estando muy cerca de la formulación de Gauss de 1821 ([Fisher, 1922](https://www.jstor.org/stable/2341124?origin=crossref&seq=1)).

Las regresiones en los años 1950 - 1970, la regresión se aplicó usando calculadoras electromecánicas, en diversas áreas de la investigación, y cuyas resoluciones podrían demorar hasta 24 horas o más, dependiendo de la magnitud de los datos a modelar. Hoy en día, el desarrollo computacional unido a los aportes de la estadística y la modelación matemática y numérica ha permitido un uso importante del análisis de regresión en diversos campos de la ciencia, la ingeniería, entre otros, siendo un área muy activa en la investigación; y nuevos métodos de regresión que implica respuestas correlacioandas (como series de tiempo y curvas de crecimiento), involucrando como datos para dicha correlación, curvas, imágenes, gráficos, objetvos de datos complejos, así como métodos de regresión que aceptan datos faltantes, entre otras. Esto ha desarrollado enteonces, una análisis de regresión denominada **regresión robusta**.

### 4.7 ESTUDIO DE CORRELACIÓN PEARSON, SPEARMAN Y KENDALL

En razón a que Pandas tiene limitación en el cálculo de la significancia estadística a pesar de poder calcular la correlación entre dos series de columnas de la base de datos, se utilizará la librería Pingouin, usando la función corr(), para obtener además del coeficiente de correlación, también su significancia, intervalo de confianza y poder estadístico entre otros ([Amat, 2020](https://www.cienciadedatos.net/documentos/pystats05-correlacion-lineal-python.html)). A manera de ejemplo, se hace con respecto a 'pm2.5' y 'Iws'.

```python
## Cálculo de correlación, significancia e intervalos con pingouin
display(pg.corr(datos['Iws'], datos['pm2.5'], method='pearson'))
display(pg.corr(datos['Iws'], datos['pm2.5'], method='spearman'))
display(pg.corr(datos['Iws'], datos['pm2.5'], method='kendall'))
```

Puede mencionarse que para el caso anterior hay una bajo coeficiente de correlación lineal, ya que se pudo observar que su tendencia no es lineal, para lo cual se sugiere un modelo de regresión no lineal. A manera de ejemplo se plantea una función polinómica de la forma $$Y = A + B X^3 + C X^5 $$:

```python
## Código para trazar el gráfico y comparar 'pm2.5' con 'Iws'
# Gráfico
fig, ax = plt.subplots(1, 1, figsize=(6,4))
x=datos['Iws']
y=datos['pm2.5']
ax.scatter(x, y, alpha= 0.8)
ax.set_xlabel('Velocidad del viento acumulada [m/s]')
ax.set_ylabel('Material Particulado PM2.5');

## Generando el modelo:
##definimos el modelo
def modelo(x, A, B, C):
    Y = A + (B*(x**3)) + (C*(x**5))
    return Y

## Buscamos los mejores parámetros del modelo, normalizando el creado
x_data = x/max(x)
y_data = y/max(y)

## Buscamos los mejores parámetros utilizando curve_fit que utiliza cuadrados mínimos no lineales
from scipy.optimize import curve_fit
popt, pcov = curve_fit(modelo, x_data, y_data)

## imprimiendo los parámetros finales
print(" A = %f, B = %f, C = %f" % (popt[0], popt[1], popt[2]) )
```

```python
x = np.linspace(0, 600, 40000)
x = x/max(x)
plt.figure(figsize=(8,5))
Y = modelo(x, *popt)
plt.plot(x_data, y_data, 'ro', label='reportado')
plt.plot(x, Y, 'bo', label='predicción')
plt.ylabel('Material Particulado PM2.5')
plt.xlabel('Velocidad del viento acumulada [m/s]')
plt.show();
```

Se sugiere una exploración de otros modelos no lineales que permiten un mejor ajuste.
